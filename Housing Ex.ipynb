## Housing Example

```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(color_codes=True)
%matplotlib inline
```

```
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data', header=-1, delim_whitespace=True)
```

```
df.head()
```

```
col_name = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
```

```
df.columns = col_name
```

```
df.describe()
```

```
print(df.info())
```

```
sns.pairplot(df, size=1.2)
```
